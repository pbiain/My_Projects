# LAB | NormalObjects - Creative Complaint Handler (LangChain)
## Step 1: Setup and Project Structure

```python
# Cell 1 - Step 1: Imports and environment setup
import os
import random
from typing import List, Dict
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI
from langchain_core.tools import tool
from langchain_core.messages import HumanMessage
from langgraph.prebuilt import create_react_agent

load_dotenv()  # Load OPENAI_API_KEY from .env

# Initialize LLM
llm = ChatOpenAI(model="gpt-4o-mini", temperature=0.7)

print("✅ Environment ready.")
```

## Step 2: Create Creative Tools

```python
# Cell 2 - Step 2: Define the four themed tools

@tool
def consult_demogorgon(complaint: str) -> str:
    """Get the Demogorgon's perspective on a complaint about the Upside Down."""
    responses = [
        f"The Demogorgon tilts its head at '{complaint}'. Perhaps you're thinking in three dimensions?",
        f"The Demogorgon suggests the problem is temporal - things work differently in the Upside Down's time.",
        f"The Demogorgon doesn't grasp '{complaint}' - maybe consistency isn't a priority there?"
    ]
    return random.choice(responses)


@tool
def check_hawkins_records(query: str) -> str:
    """Search Hawkins historical records for information about strange occurrences."""
    records = {
        "portal": "Records show portals have opened on various dates with no clear pattern. Weather, electromagnetic activity, and unknown factors seem involved.",
        "monsters": "Historical records indicate creatures from the Upside Down behave differently based on environmental factors, time of day, and proximity to certain individuals.",
        "psychics": "Records show that psychic abilities vary greatly. Some can move objects but not see the future, others see visions but cannot move things.",
        "electricity": "Hawkins has a history of electrical anomalies. Records suggest a connection between the Upside Down and electromagnetic fields."
    }
    for key, value in records.items():
        if key in query.lower():
            return value
    return f"Records don't contain specific info about '{query}', but many unexplained events have occurred in Hawkins."


@tool
def cast_interdimensional_spell(problem: str, creativity_level: str = "medium") -> str:
    """Suggest a creative interdimensional spell to fix a problem. creativity_level can be low, medium, or high."""
    creativity_multiplier = {"low": 1, "medium": 2, "high": 3}[creativity_level]
    spells = [
        f"Chant 'Becma Becma Becma' three times while holding a Walkman to recalibrate frequencies for: {problem}",
        f"Create a salt circle with a compass in the center to stabilize: {problem}",
        f"Play 'Running Up That Hill' backwards at the exact location to fix: {problem}",
        f"Arrange a lighter, compass, and personal item in a triangle while thinking about: {problem}",
    ]
    selected = random.sample(spells, min(creativity_multiplier, len(spells)))
    return "\n".join(selected)


@tool
def gather_party_wisdom(question: str) -> str:
    """Ask the D&D party (Mike, Dustin, Lucas, Will) for their collective wisdom on a problem."""
    party_responses = {
        "portal": "Mike: 'Portals open near emotional events or EM disturbances.' Dustin: 'Linked to the Mind Flayer.'",
        "monsters": "Lucas: 'Demogorgons are territorial but opportunistic.' Will: 'They sense fear and strong emotions.'",
        "psychics": "Mike: 'El's powers are tied to her emotional state.' Dustin: 'Limited by her energy.'",
        "electricity": "Lucas: 'The Upside Down interferes with electrical systems.' Dustin: 'It creates a feedback loop.'"
    }
    for key, response in party_responses.items():
        if key in question.lower():
            return response
    return "The party huddles. Mike: 'Tough one.' Dustin: 'Need more info.' Lucas: 'Think it through.' Will: 'Consult other sources?'"


tools = [consult_demogorgon, check_hawkins_records, cast_interdimensional_spell, gather_party_wisdom]

print(f"Created {len(tools)} creative tools:")
for t in tools:
    print(f"  - {t.name}: {t.description[:60]}...")
```

## Step 3: Create Agent with Tools

```python
# Cell 3 - Step 3: Build the agent
# create_react_agent (langgraph) replaces the old AgentExecutor + create_openai_tools_agent pattern

system_prompt = (
    "You are a creative problem-solving agent operating in a strange universe. "
    "You may consult unusual sources, historical records, or strange creatures. "
    "Use tools when helpful, chain multiple tools if needed. "
    "Be concise but imaginative. If no tool fits, respond creatively on your own."
)

agent_executor = create_react_agent(llm, tools, prompt=system_prompt)

print("✅ Agent created successfully.")
```

## Step 4: Test with Sample Complaints

```python
# Cell 4 - Step 4: Handle complaints using agent_executor

complaints = [
    "Why do demogorgons sometimes eat people and sometimes don't?",
    "The portal opens on different days, is there a schedule?",
    "Why can some psychics see the Downside Up and others can't?",
    "Why do creatures and power lines react so strangely together?",
]

def handle_complaint(complaint: str) -> str:
    """Handle a single complaint using the agent_executor."""
    print(f"\n{'='*60}")
    print(f"COMPLAINT: {complaint}")
    print(f"{'='*60}\n")
    result = agent_executor.invoke({"messages": [HumanMessage(content=complaint)]})
    return result["messages"][-1].content

# Test with first 2 complaints
print("Testing agent with sample complaints...\n")
for complaint in complaints[:2]:
    response = handle_complaint(complaint)
    print(f"\nRESPONSE: {response}\n")
```

## Step 5: Analyze Agent Behavior

```python
# Cell 5 - Step 5: Track and analyze tool usage

class ToolUsageTracker:
    """Track tool usage for analysis."""
    def __init__(self):
        self.usage_count = {t.name: 0 for t in tools}
        self.tool_sequences = []

    def track_from_result(self, result: dict):
        """Parse tool calls from agent result messages."""
        for msg in result.get("messages", []):
            # ToolMessages have a 'name' attribute with the tool that was called
            if hasattr(msg, "name") and msg.name in self.usage_count:
                self.usage_count[msg.name] += 1
                self.tool_sequences.append(msg.name)

    def get_statistics(self) -> Dict:
        """Get usage statistics."""
        return {
            "total_tool_calls": sum(self.usage_count.values()),
            "tool_counts": self.usage_count,
            "most_used": max(self.usage_count.items(), key=lambda x: x[1])[0] if any(self.usage_count.values()) else None,
            "tool_sequences": self.tool_sequences
        }


tracker = ToolUsageTracker()

print("Running all complaints for analysis...\n")
for complaint in complaints:
    result = agent_executor.invoke({"messages": [HumanMessage(content=complaint)]})
    tracker.track_from_result(result)
    print(f"Handled: {complaint[:50]}...")

print("\n=== Tool Usage Analysis ===")
stats = tracker.get_statistics()
print(f"Total tool calls: {stats['total_tool_calls']}")
print(f"Tool usage counts: {stats['tool_counts']}")
print(f"Most used tool: {stats['most_used']}")
print(f"Tool call sequence: {' -> '.join(stats['tool_sequences'])}")
```
