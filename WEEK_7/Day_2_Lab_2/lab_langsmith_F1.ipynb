```python
# ── Dataset config ────────────────────────────────────────────────
DATASET_NAME = "f1-knowledge-qa-v1"
DATASET_DESC = (
    "Custom F1 knowledge Q&A dataset with 15 examples across 4 categories: "
    "rules & regulations, history, teams & drivers, and race strategy. "
    "Used for evaluating LLM factual accuracy on Formula 1 domain knowledge."
)

# ── Check if dataset already exists (avoid duplicates) ────────────
existing = [ds for ds in ls_client.list_datasets() if ds.name == DATASET_NAME]

if existing:
    dataset = existing[0]
    print(f"⚠️  Dataset already exists — reusing it")
else:
    dataset = ls_client.create_dataset(
        dataset_name=DATASET_NAME,
        description=DATASET_DESC
    )
    print(f"✅ Dataset created: {dataset.name}")

print(f"   ID: {dataset.id}")

# ── Upload examples ───────────────────────────────────────────────
ls_client.create_examples(
    dataset_id=dataset.id,
    inputs=[ex["input"] for ex in f1_examples],
    outputs=[ex["output"] for ex in f1_examples],
)

print(f"✅ {len(f1_examples)} examples uploaded to LangSmith")
print(f"   Dataset: {DATASET_NAME}")
```

```python
@traceable
def f1_target(inputs: dict) -> dict:
    response = oai_client.chat.completions.create(
        model="gpt-4o-mini",
        temperature=0,
        messages=[
            {
                "role": "system",
                "content": (
                    "You are an expert Formula 1 analyst with deep knowledge of F1 history, "
                    "rules and regulations, teams, drivers, and race strategy. "
                    "Answer questions accurately and concisely. "
                    "Stick to facts — do not speculate or hallucinate statistics."
                )
            },
            {
                "role": "user",
                "content": inputs["question"]
            }
        ]
    )
    return {"answer": response.choices[0].message.content}
```

```python
test_questions = [
    {"question": "What does DRS stand for and when is a driver allowed to use it?"},
    {"question": "What does 'boxing' mean in Formula 1 team radio communication?"}
]

for q in test_questions:
    result = f1_target(q)
    print(f"Q: {q['question'][:60]}...")
    print(f"A: {result['answer'][:150]}...")
    print()
```

```python
from openevals.llm import create_llm_as_judge
from openevals.prompts import CORRECTNESS_PROMPT

def correctness_evaluator(inputs: dict, outputs: dict, reference_outputs: dict):
    evaluator = create_llm_as_judge(
        prompt=CORRECTNESS_PROMPT,
        model="openai:gpt-4o-mini",
        feedback_key="correctness",
    )
    return evaluator(
        inputs=inputs,
        outputs=outputs,
        reference_outputs=reference_outputs
    )

print("✅ Correctness evaluator defined")
```

```python
experiment_results = ls_client.evaluate(
    f1_target,
    data=DATASET_NAME,
    evaluators=[correctness_evaluator],
    experiment_prefix="f1-gpt4o-mini",
    max_concurrency=2,
)

print("✅ Evaluation complete")
print(f"   Experiment : f1-gpt4o-mini")
print(f"   Examples   : {len(list(experiment_results))}")
```

```python
import pandas as pd

results_data = []

for result in experiment_results._results:
    inputs     = result.get("run", {}).inputs or {}
    outputs    = result.get("run", {}).outputs or {}
    evaluation = result.get("evaluation_results", {})

    question = inputs.get("question") or inputs.get("inputs", {}).get("question", "")
    answer   = outputs.get("answer", "")[:100]

    correctness = None
    for eval_result in evaluation.get("results", []):
        if eval_result.key == "correctness":
            correctness = eval_result.score
            break

    results_data.append({
        "input"      : question,
        "output"     : answer,
        "correctness": correctness,
    })

df = pd.DataFrame(results_data)
print(f"✅ Retrieved {len(df)} results")
print(f"\n   Mean correctness : {df['correctness'].mean():.3f}")
print(f"   Min              : {df['correctness'].min():.3f}")
print(f"   Max              : {df['correctness'].max():.3f}")
print(f"\n{df[['input','correctness']].to_string()}")
```

```python
# Add category metadata back in
categories = [ex["metadata"]["category"] for ex in f1_examples]
difficulties = [ex["metadata"]["difficulty"] for ex in f1_examples]

# Align with df order using the question as key
category_map  = {ex["input"]["question"]: ex["metadata"]["category"]  for ex in f1_examples}
difficulty_map = {ex["input"]["question"]: ex["metadata"]["difficulty"] for ex in f1_examples}

df["category"]   = df["input"].map(category_map)
df["difficulty"]  = df["input"].map(difficulty_map)
df["correct_bin"] = df["correctness"].apply(lambda x: 1 if x else 0)

print("── Performance by category ──────────────────")
print(df.groupby("category")["correct_bin"].agg(["sum","count","mean"]).rename(
    columns={"sum":"correct","count":"total","mean":"accuracy"}
).to_string())

print("\n── Performance by difficulty ────────────────")
print(df.groupby("difficulty")["correct_bin"].agg(["sum","count","mean"]).rename(
    columns={"sum":"correct","count":"total","mean":"accuracy"}
).to_string())

print("\n── Failed examples ──────────────────────────")
failed = df[df["correct_bin"] == 0]
for _, row in failed.iterrows():
    print(f"  [{row['category']}] [{row['difficulty']}] {row['input'][:80]}")
```

```python
F1_TERMINOLOGY_PROMPT = """You are an expert Formula 1 analyst evaluating whether an answer uses accurate and precise F1 terminology.

You will be given:
- A question about Formula 1
- The model's answer

Evaluate ONLY whether the answer uses correct, domain-specific F1 terminology and avoids vague or incorrect technical language. Do NOT evaluate factual correctness — focus purely on terminology precision.

Score the answer on a scale of 0 to 1:
- 1.0: Answer uses precise F1 terminology throughout (e.g. correct names for systems, regulations, roles)
- 0.5: Answer is mostly correct in terminology but uses some vague or informal language
- 0.0: Answer uses incorrect terminology, confuses F1-specific concepts, or is too generic

Question: {input}
Answer: {output}

Respond in this exact JSON format:
{{
  "score": <float between 0 and 1>,
  "reasoning": "<one sentence explaining your score>"
}}
"""

def f1_terminology_evaluator(inputs: dict, outputs: dict, reference_outputs: dict):
    question = inputs.get("question", "")
    answer   = outputs.get("answer", "")

    response = oai_client.chat.completions.create(
        model="gpt-4o-mini",
        temperature=0,
        messages=[
            {
                "role": "user",
                "content": F1_TERMINOLOGY_PROMPT.format(
                    input=question,
                    output=answer
                )
            }
        ]
    )

    import json
    raw = response.choices[0].message.content.strip()
    parsed = json.loads(raw)

    return {
        "key"      : "f1_terminology",
        "score"    : parsed["score"],
        "comment"  : parsed["reasoning"]
    }

print("✅ F1 terminology evaluator defined")
```

```python
test_cases = [
    {
        "inputs"           : {"question": "What is an undercut strategy in Formula 1 and when is it most effective?"},
        "outputs"          : {"answer": f1_target({"question": "What is an undercut strategy in Formula 1 and when is it most effective?"})["answer"]},
        "reference_outputs": {}
    },
    {
        "inputs"           : {"question": "What does DRS stand for and when is a driver allowed to use it?"},
        "outputs"          : {"answer": f1_target({"question": "What does DRS stand for and when is a driver allowed to use it?"})["answer"]},
        "reference_outputs": {}
    },
]

for tc in test_cases:
    result = f1_terminology_evaluator(tc["inputs"], tc["outputs"], tc["reference_outputs"])
    print(f"Q      : {tc['inputs']['question'][:70]}...")
    print(f"Score  : {result['score']}")
    print(f"Comment: {result['comment']}")
    print()
```

```python
experiment_results_custom = ls_client.evaluate(
    f1_target,
    data=DATASET_NAME,
    evaluators=[correctness_evaluator, f1_terminology_evaluator],
    experiment_prefix="f1-gpt4o-mini-dual-eval",
    max_concurrency=2,
)

print("✅ Dual evaluation complete")
```

```python
dual_data = []

for result in experiment_results_custom._results:
    inputs     = result.get("run", {}).inputs or {}
    outputs    = result.get("run", {}).outputs or {}
    evaluation = result.get("evaluation_results", {})

    question = inputs.get("question") or inputs.get("inputs", {}).get("question", "")

    correctness = None
    terminology = None
    for eval_result in evaluation.get("results", []):
        if eval_result.key == "correctness":
            correctness = float(eval_result.score) if eval_result.score is not None else None
        if eval_result.key == "f1_terminology":
            terminology = float(eval_result.score) if eval_result.score is not None else None

    dual_data.append({
        "input"      : question[:70],
        "correctness": correctness,
        "terminology": terminology,
    })

df_dual = pd.DataFrame(dual_data)

print("── Dual Evaluator Results ───────────────────────────────────────")
print(df_dual.to_string())
print(f"\n── Averages ─────────────────────────────────────────────────────")
print(f"   Correctness mean : {df_dual['correctness'].mean():.3f}")
print(f"   Terminology mean : {df_dual['terminology'].mean():.3f}")
print(f"\n── Divergence (terminology=1 but correctness=0) ─────────────────")
divergent = df_dual[(df_dual["terminology"] >= 0.5) & (df_dual["correctness"] == 0)]
print(divergent[["input","correctness","terminology"]].to_string())
```

```python
@traceable
def f1_target_gpt4o(inputs: dict) -> dict:
    response = oai_client.chat.completions.create(
        model="gpt-4o",
        temperature=0,
        messages=[
            {
                "role": "system",
                "content": (
                    "You are an expert Formula 1 analyst with deep knowledge of F1 history, "
                    "rules and regulations, teams, drivers, and race strategy. "
                    "Answer questions accurately and concisely. "
                    "Stick to facts — do not speculate or hallucinate statistics."
                )
            },
            {
                "role": "user",
                "content": inputs["question"]
            }
        ]
    )
    return {"answer": response.choices[0].message.content}

print("✅ gpt-4o target function defined")
```

```python
experiment_results_gpt4o = ls_client.evaluate(
    f1_target_gpt4o,
    data=DATASET_NAME,
    evaluators=[correctness_evaluator, f1_terminology_evaluator],
    experiment_prefix="f1-gpt4o",
    max_concurrency=2,
)

print("✅ gpt-4o evaluation complete")
```

```python
def extract_results(experiment_results, model_name):
    data = []
    for result in experiment_results._results:
        inputs     = result.get("run", {}).inputs or {}
        evaluation = result.get("evaluation_results", {})

        question = inputs.get("question") or inputs.get("inputs", {}).get("question", "")

        correctness = None
        terminology = None
        for eval_result in evaluation.get("results", []):
            if eval_result.key == "correctness":
                correctness = float(eval_result.score) if eval_result.score is not None else None
            if eval_result.key == "f1_terminology":
                terminology = float(eval_result.score) if eval_result.score is not None else None

        data.append({
            "model"      : model_name,
            "input"      : question[:70],
            "correctness": correctness,
            "terminology": terminology,
        })
    return pd.DataFrame(data)

df_mini  = extract_results(experiment_results_custom, "gpt-4o-mini")
df_4o    = extract_results(experiment_results_gpt4o,  "gpt-4o")

# ── Summary comparison ────────────────────────────────────────────
summary = pd.DataFrame([
    {
        "model"           : "gpt-4o-mini",
        "correctness_mean": df_mini["correctness"].mean(),
        "terminology_mean": df_mini["terminology"].mean(),
        "correct_count"   : int(df_mini["correctness"].sum()),
        "cost_per_1k_input_tokens"  : 0.15,
        "cost_per_1k_output_tokens" : 0.60,
    },
    {
        "model"           : "gpt-4o",
        "correctness_mean": df_4o["correctness"].mean(),
        "terminology_mean": df_4o["terminology"].mean(),
        "correct_count"   : int(df_4o["correctness"].sum()),
        "cost_per_1k_input_tokens"  : 2.50,
        "cost_per_1k_output_tokens" : 10.00,
    },
])

print("── A/B Summary ──────────────────────────────────────────────────")
print(summary.to_string(index=False))

print("\n── Per-example comparison ───────────────────────────────────────")
df_compare = df_mini[["input","correctness","terminology"]].copy()
df_compare.columns = ["input", "mini_correctness", "mini_terminology"]
df_compare["4o_correctness"] = df_4o["correctness"].values
df_compare["4o_terminology"] = df_4o["terminology"].values
df_compare["delta"] = df_compare["4o_correctness"] - df_compare["mini_correctness"]
print(df_compare.to_string())

print(f"\n── Examples where gpt-4o outperforms gpt-4o-mini ────────────────")
better = df_compare[df_compare["delta"] > 0]
if len(better) == 0:
    print("   None — both models performed identically on these examples")
else:
    print(better[["input","mini_correctness","4o_correctness"]].to_string())
```
